# About dataset

This is a dataset about songs that made entries on the 2016 spotify top 200 chart containing information on audio features of each track, keys, mode, time signature, artists and sources

# Data analysis on the 2016 spotify  top 200 chart using Python 

```python
#Importing necessary libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
#data loading
df = pd.read_csv("C:/Users/HP/Desktop/Projects/Datasets/spotify_top_songs_audio_features.csv")
df.head()
```

```python
#Checking core data informations
df.info()
```

```python
#Checking for missing columns
df.isna().sum().to_frame()
```

```python
#Dropping unneeded column
df.drop("id", axis = 1, inplace=True)
```

## Making sure all Categorical columns are assigns to the appropriate data type 

```python
df["key"].unique()
```

```python
df["key"] = df["key"].astype("category")
```

```python
df["mode"].unique()
```

```python
df["mode"] = df["mode"].astype("category")
```

```python
df["time_signature"].unique()
```

```python
signature_changes = {
    "1" : "1 beat"
}
df["time_signature"] = df["time_signature"].replace(signature_changes)
```

```python
df["time_signature"].unique()
```

```python
df["time_signature"] = df["time_signature"].astype("category")
```

```python
df.info()
```
